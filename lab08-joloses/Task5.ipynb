### 5.1. Create a ["Ridgeline"](https://seaborn.pydata.org/examples/kde_ridgeplot.html) plot for the plots from 3.1.2 and 3.1.3 .


***Important***: When referring to Type of the pokemon, please use just the **"Type 1"** Column.

<img src="images/ridgeline.png" width="300px">

```python
# Your Solution Here.

z = sns.FacetGrid(df, row="Type 1", hue="Type 1", aspect=4, height=1)

z.map(sns.kdeplot, "Weighted Score", bw_adjust=.5, clip_on=False, fill=True, alpha=1, linewidth=1.5)
z.map(sns.kdeplot, "Weighted Score", clip_on=False, color="w", lw=2, bw_adjust=.5)


def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color, ha="left", va="center", transform=ax.transAxes)


z.map(label, "Weighted Score")

z.set_titles("")
z.set(yticks=[], ylabel="")
z.despine(bottom=True, left=True)
```
