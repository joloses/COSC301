#### 2.5.1. Drop the 'Generation', 'Sp. Atk', 'Sp. Def', 'Total', and the '#' columns

```python
# Your solution here
df2 = df.drop(['Generation', 'Sp. Atk','Sp. Def','Total','#'], axis = 1)
df2
```

#### 2.5.2. Drop any NaN values in HP, Attack, Defense, Speed

```python
# Your solution here
df2 = df2.dropna(subset=['HP', 'Attack', 'Defense', 'Speed'], inplace=True)
df2
```

#### 2.5.3. Reset the index to get a new index without missing values

```python
# Your solution here
df = df.reset_index()
df
```

#### 2.5.4. A new column was added called `index`; remove it. 

```python
# Your solution here
df = df.drop(['index'], axis = 1)
df
```

#### 2.5.5. Calculate a new column called "Weighted Score" that computes an aggregate score comprising:

- 20% 'HP'
- 40% 'Attack'
- 30% 'Defense'
- 10% 'Speed'


```python
# Your solution here
df['Weighted Score'] = (df['HP'] * 0.2) + (df['Attack'] * 0.4) + (df['Defense'] * 0.3) + (df['Speed'] * 0.1)
df
```
